```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
service = Service()

options = webdriver.ChromeOptions()

driver = webdriver.Chrome(service=service,
                           options=options)
```

```python
url = 'https://www.sofascore.com/tournament/football/brazil/brasileirao-serie-a/325#id:48982'

driver.get(url)
```

```python
elements = driver.find_elements(By.CLASS_NAME, 'sc-631ef08b-0')
hrefs = [element.get_attribute('href') for element in elements]
```

```python
hrefs
```

```python
with open('brasileirao_2023_links.txt', 'w') as file:
    for href in hrefs:
        if href:  # Verifica se o href não é None ou uma string vazia
            file.write(href + '\n')
```

```python
import re

# Sua string original
url = 'https://www.sofascore.com/flamengo-sao-paulo/GOsGuc#id:11067620'

# Nome do time que você quer remover
time = 'flamengo'

# Expressão regular para capturar a parte após "sofscore.com/" até a próxima "/"
match = re.search(r'sofascore\.com/([^/]+)', url)

if match:
    resultado = match.group(1)
    print(f"Original: {resultado}")

    # Removendo o time especificado da string resultado
    if time in resultado:
        resultado = resultado.replace(time, '').replace('--', '-')
        # Remover o traço inicial ou final, caso existam
        resultado = resultado.strip('-')
    
    print(f"Modificado: {resultado}")
else:
    print("Partida não encontrada.")
```
